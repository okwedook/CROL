```python
from CloudEnv import CloudEnv # import custom environment
from CloudEnv import STOP, dtype, n, m
from gym.spaces import Box
import numpy as np
```

```python
class NewCloudEnv(CloudEnv):
    def __init__(self):
        super().__init__()
        self.observation_space = Box(low=0, high=STOP, shape=(n, m, 5), dtype=dtype)

    def get_state(self):
        state = np.zeros(shape=(n, m, 5))
        for i in range(n):
            for j in range(m):
                state[i][j] = [
                    self.vms[i].cpu,
                    self.vms[i].mem,
                    self.servers[j].cpu,
                    self.servers[j].mem,
                    int(self.vms[i] <= self.servers[j])
                ]
        return state


env = NewCloudEnv()
# env = CloudEnv()
```

```python
from stable_baselines3 import A2C, TD3
from stable_baselines3.common.sb2_compat.rmsprop_tf_like import RMSpropTFLike
import os
```

```python
log_path = os.path.join('Training', 'Logs')
model = A2C(
    policy='MlpPolicy',
    env=env,
    learning_rate=0.000005,
#     gamma=0.97,
    policy_kwargs=dict(optimizer_class=RMSpropTFLike),
    verbose=True,
    tensorboard_log=log_path,
#     n_steps=5
)
```

```python
model.learn(total_timesteps=1000000)
```
